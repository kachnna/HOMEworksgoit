# Podstawowa sieć konwolucyjna (CNN)

```python
import keras
import tensorflow as tf
from tensorflow.keras import layers, models, datasets
import numpy as np
import matplotlib.pyplot as plt

# Załadowanie danych
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalizacja danych
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Dodanie kanału koloru (jeden kanał dla obrazów grayscale)
x_train = x_train[..., np.newaxis]
x_test = x_test[..., np.newaxis]

# Tworzenie modelu CNN
model_cnn = models.Sequential([
    layers.Input(shape=(28, 28, 1)),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

# Kompilacja modelu
model_cnn.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Trening modelu
history_cnn = model_cnn.fit(
    x_train, y_train,
    epochs=30,
    batch_size=256,
    validation_data=(x_test, y_test),
    verbose=2
)

# Ocena modelu na danych testowych
test_loss_cnn, test_accuracy_cnn = model_cnn.evaluate(x_test, y_test, verbose=2)
print(f"Dokładność na zbiorze testowym (CNN): {test_accuracy_cnn:.4f}")

# Wykres historii treningu
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_cnn.history['loss'], label='loss')
plt.plot(history_cnn.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_cnn.history['accuracy'], label='accuracy')
plt.plot(history_cnn.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend()

plt.show()
```

# CNN z VGG16 jako bazą konwolucyjną

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Dropout, Input, Conv2D, MaxPooling2D
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import tensorflow as tf

# Załadowanie danych
(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

# Normalizacja danych
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# Dodanie kanału koloru (jeden kanał dla obrazów grayscale)
x_train = np.repeat(x_train[..., np.newaxis], 3, -1)
x_test = np.repeat(x_test[..., np.newaxis], 3, -1)

# Przeskalowanie obrazów do 32x32 pikseli
x_train_resized = tf.image.resize(x_train, [32, 32])
x_test_resized = tf.image.resize(x_test, [32, 32])

# Załadowanie VGG16 bez w pełni połączonych warstw, z wagami z ImageNet
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(32, 32, 3))

# Dodanie własnych warstw na górę sieci VGG16
x = base_model.output
x = Flatten()(x)
x = Dense(256, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(10, activation='softmax')(x)

# Tworzenie końcowego modelu
model_vgg16 = Model(inputs=base_model.input, outputs=predictions)

# Zablokowanie wszystkich warstw VGG16 podczas wstępnego uczenia
for layer in base_model.layers:
    layer.trainable = False

# Kompilacja modelu
model_vgg16.compile(
    optimizer=Adam(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Wstępne uczenie modelu
history_vgg16 = model_vgg16.fit(
    x_train_resized, y_train,
    epochs=10,
    batch_size=256,
    validation_data=(x_test_resized, y_test),
    verbose=2
)

# Odblokowanie ostatnich warstw VGG16 do fine-tuningu
for layer in base_model.layers[-4:]:
    layer.trainable = True

# Kompilacja modelu po odblokowaniu warstw
model_vgg16.compile(
    optimizer=Adam(1e-5),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Fine-tuning modelu
history_vgg16_ft = model_vgg16.fit(
    x_train_resized, y_train,
    epochs=20,
    batch_size=256,
    validation_data=(x_test_resized, y_test),
    verbose=2
)

# Ocena modelu na danych testowych
test_loss_vgg16, test_accuracy_vgg16 = model_vgg16.evaluate(x_test_resized, y_test, verbose=2)
print(f"Dokładność na zbiorze testowym (VGG16): {test_accuracy_vgg16:.4f}")

# Wykres historii treningu
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_vgg16.history['loss'] + history_vgg16_ft.history['loss'], label='loss')
plt.plot(history_vgg16.history['val_loss'] + history_vgg16_ft.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_vgg16.history['accuracy'] + history_vgg16_ft.history['accuracy'], label='accuracy')
plt.plot(history_vgg16.history['val_accuracy'] + history_vgg16_ft.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy')
plt.legend()

plt.show()
```
