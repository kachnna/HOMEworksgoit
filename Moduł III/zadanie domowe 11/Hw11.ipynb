```python
import matplotlib.pyplot as plt
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, SimpleRNN, LSTM, Bidirectional, Dense
from keras.optimizers import RMSprop
```

Ustawienie hiperparametrów:

```python
max_features = 10000# liczba słów traktowanych jako cechy
maxlen = 500# ograniczenie liczby słów w tekście
batch_size = 32
```

Pobieranie danych

```python
(input_train, y_train), (input_test, y_test) = imdb.load_data(num_words=max_features)
```

Wstępne przetwarzanie danych szeregowych:

```python
input_train = pad_sequences(input_train, maxlen=maxlen)
input_test = pad_sequences(input_test, maxlen=maxlen)
```

```python
def train_rnn_model(model, model_name):
    model.compile(optimizer=RMSprop(), loss='binary_crossentropy', metrics=['acc'])
    history = model.fit(input_train, y_train, epochs=5, batch_size=batch_size, validation_split=0.3)
    
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    
    epochs = range(1, len(acc) + 1)
    
    plt.plot(epochs, acc, 'bo', label='Training accuracy')
    plt.plot(epochs, val_acc, 'b', label='Validation accuracy')
    plt.title(f'{model_name}: Training and validation accuracy')
    plt.legend()
    
    plt.figure()
    
    plt.plot(epochs, loss, 'bo', label='Training loss')
    plt.plot(epochs, val_loss, 'b', label='Validation loss')
    plt.title(f'{model_name}: Training and validation loss')
    plt.legend()
    
    plt.show()
    
    # Evaluate the model on test data
    test_loss, test_acc = model.evaluate(input_test, y_test)
    print(f'{model_name} - Test loss:', test_loss)
    print(f'{model_name} - Test accuracy:', test_acc)
    
    return test_acc

# SimpleRNN model
model_simple_rnn = Sequential()
model_simple_rnn.add(Embedding(max_features, 32))
model_simple_rnn.add(SimpleRNN(32))
model_simple_rnn.add(Dense(1, activation='sigmoid'))

# Deep RNN model
model_deep_rnn = Sequential()
model_deep_rnn.add(Embedding(max_features, 32))
model_deep_rnn.add(SimpleRNN(32, return_sequences=True))
model_deep_rnn.add(SimpleRNN(32))
model_deep_rnn.add(Dense(1, activation='sigmoid'))

# LSTM model
model_lstm = Sequential()
model_lstm.add(Embedding(max_features, 32))
model_lstm.add(LSTM(32))
model_lstm.add(Dense(1, activation='sigmoid'))

# Deep LSTM model
model_deep_lstm = Sequential()
model_deep_lstm.add(Embedding(max_features, 32))
model_deep_lstm.add(LSTM(32, return_sequences=True))
model_deep_lstm.add(LSTM(32))
model_deep_lstm.add(Dense(1, activation='sigmoid'))

# Bidirectional LSTM model
model_bilstm = Sequential()
model_bilstm.add(Embedding(max_features, 32))
model_bilstm.add(Bidirectional(LSTM(32)))
model_bilstm.add(Dense(1, activation='sigmoid'))

# Bidirectional SimpleRNN model
model_birnn = Sequential()
model_birnn.add(Embedding(max_features, 32))
model_birnn.add(Bidirectional(SimpleRNN(32)))
model_birnn.add(Dense(1, activation='sigmoid'))

# Train and evaluate models
acc_simple_rnn = train_rnn_model(model_simple_rnn, "SimpleRNN")
acc_deep_rnn = train_rnn_model(model_deep_rnn, "Deep RNN")
acc_lstm = train_rnn_model(model_lstm, "LSTM")
acc_deep_lstm = train_rnn_model(model_deep_lstm, "Deep LSTM")
acc_bilstm = train_rnn_model(model_bilstm, "Bidirectional LSTM")
acc_birnn = train_rnn_model(model_birnn, "Bidirectional RNN")

# Comparing the results
print("Accuracy Comparison:")
print(f"SimpleRNN Accuracy: {acc_simple_rnn}")
print(f"Deep RNN Accuracy: {acc_deep_rnn}")
print(f"LSTM Accuracy: {acc_lstm}")
print(f"Deep LSTM Accuracy: {acc_deep_lstm}")
print(f"Bidirectional LSTM Accuracy: {acc_bilstm}")
print(f"Bidirectional RNN Accuracy: {acc_birnn}")
```

Podsumowanie

```python
print("\n\nModel Accuracies:")
print(f"- Simple RNN: {acc_simple_rnn}")
print(f"- Deep RNN: {acc_deep_rnn}")
print(f"- Simple LSTM: {acc_lstm}")
print(f"- Deep LSTM: {acc_deep_lstm}")
print(f"- Bidirectional LSTM: {acc_bilstm}")
print(f"- Bidirectional RNN: {acc_birnn}")
print("\n\nModel Details below.")
print("\n\nSIMPLE RNN MODEL:")
model_simple_rnn.summary()
print("\nDEEP RNN MODEL:")
model_deep_rnn.summary()
print("\nSIMPLE LSTM MODEL:")
model_lstm.summary()
print("\nDEEP LSTM MODEL:")
model_deep_lstm.summary()
print("\nBIDIRECTIONAL LSTM MODEL:")
model_bilstm.summary()
print("\nBIDIRECTIONAL RNN MODEL:")
model_birnn.summary()
```
